## Мини-теория (кратко, через пример)

### 1) Что такое программа
Программа — это набор команд. В Python команды выполняются сверху вниз.

```python
print("Hello, AI Diploma!")
```

### 2) Переменные
Переменная — это “коробка” с именем, где лежит значение:

```python
course = "grAI621"
year = 2026
print(course, year)
```

### 3) Зачем функции в курсе
Мы будем писать решения через функции, потому что:
- их легко проверять через `assert`
- их удобно переносить в VS Code (в диплом)

### 4) Идея репозитория и коммита (очень просто)
Репозиторий — папка проекта + история изменений.  
Коммит — “снимок” состояния проекта.
