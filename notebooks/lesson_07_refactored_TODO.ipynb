```python
from typing import Any
```

```python
# Функционал: вычисляет сумму всех целых чисел от 1 до n включительно.
# Алгоритм: создаём накопитель total, циклом for проходим по диапазону 1..n и добавляем каждое число в сумму.
def sum_1_to_n(n: int) -> int:
    total = 0
    for i in range(1, n + 1):
            total += i
    return total
```

```python
# Функционал: считает, сколько чётных чисел находится в списке nums.
# Алгоритм: перебираем элементы списка, проверяем остаток от деления на 2 и увеличиваем счётчик для чётных значений.
def count_evens(nums: list[int]) -> int:
    c = 0
    for x in nums:
        if x % 2 == 0:
            c += 1
    return c
```

```python
# Функционал: возвращает количество положительных, отрицательных и нулевых элементов списка.
# Алгоритм: заводим три счётчика, перебираем список и по условию if/elif/else увеличиваем нужный счётчик.
def sign_counts(nums: list[int]) -> dict:
    pos = neg = zero = 0
    for x in nums:
        if x > 0:
            pos += 1
        elif x < 0:
            neg += 1
        else:
            zero += 1
    return {"pos": pos, "neg": neg, "zero": zero}
```

```python
# Функционал: находит индекс первого элемента списка, который больше заданного x.
# Алгоритм: с помощью enumerate перебираем пары (индекс, значение); как только находим v > x, сразу возвращаем индекс, иначе возвращаем -1.
def find_first_gt(nums: list[int], x: int) -> int:
    for i, v in enumerate(nums):
        if v > x:
            return i
    return -1
```

```python
# Функционал: формирует новый список из квадратов только чётных чисел исходного списка.
# Алгоритм: создаём пустой список out, перебираем nums и добавляем x * x только для элементов, которые делятся на 2 без остатка.
def squares_of_evens(nums: list[int]) -> list[int]:
    out: list[int] = []
    for x in nums:
        if x % 2 == 0:
            out.append(x * x)
    return out
```

```python
# Функционал: возвращает список уникальных элементов без изменения их исходного порядка.
# Алгоритм: используем множество seen для запоминания уже встреченных элементов; если элемент новый, добавляем его и в seen, и в результат.
def unique_preserve_order(items: list[Any]) -> list[Any]:
    seen = set()
    out = []
    for i in items:
        if i not in seen:
            seen.add(i)
            out.append(i)
    return out
```

```python
# Функционал: находит первую непустую строку в списке и возвращает её без лишних пробелов.
# Алгоритм: перебираем список строк, проверяем что элемент является строкой и после strip() не пустой; при первом совпадении сразу возвращаем результат.
def first_non_empty(strings: list[str]):
    for s in strings:
        if  isinstance(s, str) and s.strip() != "":
            return s.strip()
    return None
```

```python
# Функционал: суммирует числовые значения по указанному ключу в списке словарей.
# Алгоритм: перебираем словари, получаем значение по ключу через get, проверяем что это число, и добавляем его к общей сумме.
def sum_by_key(rows: list[dict], key: str) -> float:
    total = 0.0
    for d in rows:
        val = d.get(key)
        try:
            f_val = float(val)
            total += f_val
        except ValueError, TypeError:
            pass
    return total
```

```python
from datetime import datetime
# Функционал: считает общее число элементов, количество уникальных непустых title и самую позднюю дату created_at.
# Алгоритм: проходим по списку словарей, собираем очищенные непустые заголовки в множество, а дату created_at сравниваем с текущей максимальной.
def items_stats(items: list[dict]) -> dict:
    total = len(items)
    titles = set()
    last_created_at = None
    for item in items:

        title = item.get("title")
        if title is not None and isinstance(title, str):
            title = title.strip()
            if title:
                titles.add(title)

        if last_created_at is None or datetime.fromisoformat(item.get("created_at")) > datetime.fromisoformat(last_created_at):
            last_created_at = item.get("created_at")
            
    return {"total": total, "unique_titles": len(titles), "last_created_at": last_created_at}
```

```python
# Функционал: отбирает словари, у которых поле title содержит подстроку query без учёта регистра.
# Алгоритм: переводим query в нижний регистр, затем перебираем элементы и добавляем в результат только те, где q входит в title.lower().
def filter_items_by_title(items: list[dict], query: str) -> list[dict]:
    q = str(query).lower()
    out = []
    for d in items:
        if contains_substring(d.get("title"), q):
            out.append(d)
    return out

def contains_substring(text: str, query: str) -> bool:    
    return query in text.lower()
```

```python
# Небольшая демонстрация
# Запусти после заполнения функций
print(sum_1_to_n(5))
print(count_evens([1, 2, 3, 4]))
```

```python
# =========================
# LESSON 07 — TESTS (НЕ МЕНЯТЬ)
# =========================

def run_all_tests():
    assert sum_1_to_n(0) == 0
    assert sum_1_to_n(5) == 15

    assert count_evens([1, 2, 3, 4]) == 2
    assert count_evens([]) == 0

    assert sign_counts([1, -2, 0]) == {"pos": 1, "neg": 1, "zero": 1}
    assert sign_counts([]) == {"pos": 0, "neg": 0, "zero": 0}

    assert find_first_gt([1, 2, 10], 3) == 2
    assert find_first_gt([1, 2, 3], 100) == -1
    assert find_first_gt([], 0) == -1

    assert squares_of_evens([1, 2, 3, 4]) == [4, 16]
    assert squares_of_evens([1, 3, 5]) == []
    assert squares_of_evens([0, 2]) == [0, 4]

    assert unique_preserve_order(["a", "b", "a", "c", "b"]) == ["a", "b", "c"]
    assert unique_preserve_order([]) == []

    assert first_non_empty(["  ", "x"]) == "x"
    assert first_non_empty(["", "   "]) is None

    assert sum_by_key([{"a": 1}, {"a": 2}, {"b": 5}], "a") == 3.0
    assert sum_by_key([], "a") == 0.0

    items = [
        {"title": " A ", "created_at": "2026-02-25T10:00:00"},
        {"title": "B", "created_at": "2026-02-25T12:00:00"},
        {"title": "A", "created_at": "2026-02-25T11:00:00"},
        {"title": "   ", "created_at": "2026-02-25T13:00:00"},
    ]
    st = items_stats(items)    
    assert st["total"] == 4
    assert st["unique_titles"] == 2
    assert st["last_created_at"] == "2026-02-25T13:00:00"
    assert items_stats([]) == {"total": 0, "unique_titles": 0, "last_created_at": None}

    found = filter_items_by_title(items, "a")
    assert len(found) == 2
    assert filter_items_by_title(items, "zzz") == []

    print("✅ LESSON 07: all tests passed")

run_all_tests()
```
