## Теория: термины, особенности, синтаксис

### Термины
- **Список (list)** — упорядоченная **изменяемая** коллекция объектов.
- **Индекс** — позиция элемента (с 0): `items[0]`, последний — `items[-1]`.
- **Срез (slice)** — часть списка: `items[start:end:step]`, где `end` **не включается**.
- **Итерация** — перебор элементов циклом `for`.
- **Фильтрация** — выбор элементов по условию.
- **Пагинация** — выдача элементов “страницами” (`page`, `per_page`).
- **Слияние (merge)** — объединение **двух отсортированных** списков в один отсортированный.

### Создание и доступ
```python
nums = [1, 2, 3]
first = nums[0]   # 1
last = nums[-1]   # 3
```

**Особенность:** неверный индекс → `IndexError`.

### Срезы (slice)
```python
a = [0, 1, 2, 3, 4, 5]
a[1:4]   # [1, 2, 3]
a[:3]    # [0, 1, 2]
a[3:]    # [3, 4, 5]
a[::-1]  # разворот
```
**Особенность:** `end` не входит (полуинтервал).

### Методы списков (часто)
- `append(x)` — добавить в конец
- `pop()` / `pop(i)` — удалить и вернуть
- `insert(i, x)` — вставить
- `remove(x)` — удалить по значению (первое совпадение)

### Шаблоны обработки
Фильтрация:
```python
out = [x for x in nums if x != 0]
```
Разбиение на чанки:
```python
chunk = nums[i:i+size]
```
Пагинация:
```python
start = (page-1)*per_page
end = start + per_page
page_items = items[start:end]
```
Слияние (merge) двух отсортированных списков — через два указателя `i` и `j`.

### Зачем это в дипломе
- `items` — список сущностей (dict).
- `paginate(items, page, per_page)` — выдача результатов.
- `filter_items_by_score` — фильтрация по метрике.
- `chunk` — батчи для обработки.

## Импорт и подготовка

````python
from typing import Any
````

## Функция 1: `remove_zeros`

````python
# Краткое описание функционала:
# Удаляет все нули из списка и возвращает новый список без нулевых элементов.
#
# Краткое описание алгоритма:
# Проходим по всем элементам исходного списка и оставляем только те значения, которые не равны 0.
def remove_zeros(nums:list[int]) -> list[int]:
    return [x for x in nums if x != 0]
````

## Функция 2: `reverse_list`

````python
# Краткое описание функционала:
# Разворачивает список в обратном порядке и возвращает новый список.
#
# Краткое описание алгоритма:
# Начинаем с последнего индекса, идём к первому и по одному добавляем элементы в результирующий список.
def reverse_list(nums: list[int]) -> list[int]:
    return nums[::-1]
````

## Функция 3: `chunk`

````python
# Краткое описание функционала:
# Разбивает список на части фиксированного размера и возвращает список таких частей.
#
# Краткое описание алгоритма:
# Проверяем корректность размера, затем двигаемся по списку шагом size и берём срезы нужной длины.
def chunk(lst: list[Any], size: int) -> list[list[Any]]:
    if size <= 0:
        raise ValueError("Size must be a positive integer")
    return [lst[i:i + size] for i in range(0, len(lst), size)]        
````

## Функция 4: `flatten`

````python
# Краткое описание функционала:
# Преобразует список списков в один общий плоский список.
#
# Краткое описание алгоритма:
# Перебираем каждый вложенный список, затем каждый его элемент и добавляем его в общий результат.
def flatten(chunks: list[list[Any]]) -> list[Any]:
    return [item for chunk in chunks for item in chunk]
````

## Функция 5: `merge_sorted`

````python
# Краткое описание функционала:
# Объединяет два уже отсортированных списка в один отсортированный список.
#
# Краткое описание алгоритма:
# Сравниваем текущие элементы двух списков двумя указателями, добавляем меньший и в конце дописываем остаток.
def merge_sorted(a: list[int], b: list[int]) -> list[int]:
    out = []
    i, j = 0, 0
    while i < len(a) and j < len(b):
        if a[i] < b[j]:
            out.append(a[i])
            i += 1
        else:
            out.append(b[j])
            j += 1
    out.extend(a[i:])
    out.extend(b[j:])    
    return out
            
````

## Функция 6: `unique_preserve_order`

````python
# Краткое описание функционала:
# Удаляет повторы из списка, сохраняя порядок первого появления элементов.
#
# Краткое описание алгоритма:
# Храним множество уже встреченных значений и добавляем элемент в результат только при первом появлении.
def unique_preserve_order(items: list[Any]) -> list[Any]:
    seen = set()
    out = []
    for item in items:
        if item not in seen:
            seen.add(item)
            out.append(item)
    return out
````

## Функция 7: `rotate_left`

````python
# Краткое описание функционала:
# Циклически сдвигает элементы списка влево на k позиций.
#
# Краткое описание алгоритма:
# Находим эффективный сдвиг через остаток от деления и собираем новый список из двух срезов.
def rotate_left(items: list[Any], k: int) -> list[Any]:
    if not items:
        return []
    k = k % len(items)
    return items[k:] + items[:k]
````

## Функция 8: `insert_sorted`

````python
# Краткое описание функционала:
# Вставляет число в отсортированный список так, чтобы порядок сортировки сохранился.
#
# Краткое описание алгоритма:
# Идём слева направо, ищем первую позицию, где x не больше текущего элемента, вставляем x и затем продолжаем копирование.
def insert_sorted(nums: list[int], x: int) -> list[int]:    
    n = [i for i, num in enumerate(nums) if num >= x]
    n = n[0] if n else len(nums)
    a = nums[:n] + [x] + nums[n:]
    return a
````

## Функция 9: `paginate`

````python
# Краткое описание функционала:
# Возвращает одну страницу элементов списка по номеру страницы и размеру страницы.
#
# Краткое описание алгоритма:
# Вычисляем начальный и конечный индексы страницы по формулам и возвращаем соответствующий срез списка.
def paginate(items: list[Any], page: int, per_page: int) -> list[Any]:
    if page < 0:
        raise ValueError("Page must be zero at least")
    if per_page <= 0:
        raise ValueError("Per_page must be a positive integer")
    start = (page - 1) * per_page
    end = start + per_page
    return items[start:end]
````

## Функция 10: `filter_items_by_score`

````python
# Краткое описание функционала:
# Оставляет только те словари, у которых поле score существует, является числом и не меньше порога.
#
# Краткое описание алгоритма:
# Перебираем список словарей, безопасно получаем score, проверяем тип и значение, после чего добавляем подходящие элементы в результат.
def filter_items_by_score(items: list[dict], threshold: float) -> list[dict]:
    out = []
    for item in items:
        score = item.get("score")
        if isinstance(score, (int, float)) and score >= threshold:
            out.append(item)
    return out
````
