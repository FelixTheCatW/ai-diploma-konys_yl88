
# Занятие 12 — Функции в Python

## Теория

**Функция** — это именованный блок кода, который выполняет определённую задачу.

Зачем нужны функции:
- разбивать программу на части
- переиспользовать код
- упрощать тестирование
- делать код понятным

### Синтаксис функции

```python
def имя_функции(параметры):
    код
    return результат
```

**def** — объявление функции  
**параметры** — входные данные  
**return** — возвращаемое значение

Алгоритм:
1. объявляем функцию
2. передаем параметры
3. выполняем вычисления
4. возвращаем результат
