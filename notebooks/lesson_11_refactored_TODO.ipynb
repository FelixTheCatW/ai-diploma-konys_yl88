## Теория: термины, особенности, синтаксис

### Термины
- **Кортеж (tuple)** — упорядоченная **неизменяемая** коллекция.
- **Множество (set)** — неупорядоченная коллекция **уникальных** элементов.
- **Хэшируемость** — свойство объекта быть ключом dict / элементом set.
  - Хэшируемые: `int`, `str`, `tuple` (если внутри тоже хэшируемые)
  - НЕ хэшируемые: `list`, `dict`, `set`

### Синтаксис tuple
```python
p = (10, 20)
x, y = p          # распаковка
pair = 1, 2       # тоже tuple
single = (5,)     # кортеж из 1 элемента (запятая обязательна)
```

**Особенность:** tuple нельзя менять: `p[0] = 99` → ошибка.

### Синтаксис set
```python
s = {1, 2, 2, 3}   # {1, 2, 3}
s.add(4)
2 in s             # True
```

### Операции множеств
- `a | b` — объединение (union)
- `a & b` — пересечение (intersection)
- `a - b` — разность (difference)
- `a <= b` — подмножество (subset)

### Зачем это в дипломе
- Уникальность: быстро убрать дубли (`set`).
- Теги/категории: хранить как `set[str]`.
- Координаты/пары значений: вернуть как `tuple`.
