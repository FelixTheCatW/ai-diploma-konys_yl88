### 2) f-строки — основной инструмент
```python
name = "Иван"
age = 20

```

### 3) Формат чисел (2 знака после точки)
```python
x = 12.3
print(f"{x:.2f}")   # 12.30
```

### 4) Нормализация строк
- `strip()` — убирает пробелы по краям  
- `split()` — разбивает на слова (лишние пробелы игнорируются)  
- `" ".join(...)` — склеивает обратно

```python
s = "  a   b  "
words = s.split()        # ["a", "b"]
norm = " ".join(words)   # "a b"
```
