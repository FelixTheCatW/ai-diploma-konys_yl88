
### 2) Деление на ноль
Деление на ноль вызывает ошибку. В прикладном коде часто лучше вернуть `None`:
```python
def safe_div(a, b):
    if b == 0:
        return None
    return round(a / b, 2)
```

### 3) clamp и нормализация
- `clamp(x, lo, hi)` ограничивает значение диапазоном.
- нормализация в [0,1] часто используется для метрик и масштабирования.

### 4) Медиана
Медиана — середина отсортированного списка (или среднее двух центральных).
