## Теория: термины, особенности, синтаксис

### Термины
- **Словарь (dict)** — структура данных “ключ → значение”.
- **Ключ (key)** — уникальный идентификатор внутри словаря (часто `str` или `int`).
- **Значение (value)** — любые данные: число, строка, list, dict.
- **Хэшируемый тип** — тип, который можно использовать как ключ (`str`, `int`, `tuple`…), но **не** `list` и **не** `dict`.
- **Вложенный словарь** — dict внутри dict: `d["a"]["b"]`.
- **Индекс (index)** — словарь вида `{id: item}`, ускоряет поиск.

### Синтаксис: создание/доступ/изменение
```python
item = {"id": 1, "title": "Hello", "payload": {}}
item["title"] = "Hello!"          # изменить
title = item.get("title", "")     # безопасно: если нет ключа — вернёт default
```

**Особенности**
- `item["missing"]` → `KeyError`, если ключа нет.
- `item.get("missing")` → вернёт `None` (или ваш default).
- `in` проверяет наличие ключа: `"id" in item`.

### setdefault и накопление (частоты/группировка)
```python
freq = {}
for w in ["a","b","a"]:
    freq[w] = freq.get(w, 0) + 1
# {"a": 2, "b": 1}

groups = {}
for w in ["a", "bb", "c"]:
    groups.setdefault(len(w), []).append(w)
# {1: ["a","c"], 2: ["bb"]}
```

### Вложенные данные: безопасный доступ
```python
d = {"a": {"b": {"c": 5}}}
# безопасно: если где-то нет ключа — вернуть default
```

### Зачем это в дипломе
- Сущность **item** удобно хранить как dict.
- Для быстрого поиска по id делаем индекс: `index = {item["id"]: item for item in items}`.
- Группировки и частоты — базовые операции анализа данных.

````python
# Функционал:
# Создаёт словарь-объект с ключами id, title и payload.
# Алгоритм:
# 1) Проверяет payload: если он равен None, заменяет его на пустой словарь.
# 2) Приводит item_id к int, title к str, payload копирует в новый dict.
# 3) Возвращает готовый словарь с единым форматом записи.
def make_item(item_id: int, title: str, payload: dict) -> dict:
    if payload is None:
        payload = {}
    return {"id": int(item_id), "title": str(title), "payload": dict(payload) }
````

````python
# Функционал:
# Безопасно получает значение по ключу из словаря.
# Алгоритм:
# 1) Обращается к методу get у словаря.
# 2) Если ключ найден, возвращает его значение.
# 3) Если ключа нет, возвращает default.
def safe_get(d: dict, key: str, default=None):
    if d is None or key is None or not isinstance(key, str):
        return default
    return d.get(key, default)
````

````python
# Функционал:
# Объединяет два словаря в новый словарь.
# Алгоритм:
# 1) Создаёт копию первого словаря.
# 2) Добавляет в неё пары из второго словаря через update.
# 3) Возвращает объединённый словарь.
def merge_dicts(a: dict, b: dict) -> dict:
    result = {} if a is None else a.copy()
    result.update(b)
    return result
````

````python
# Функционал:
# Считает, сколько раз каждое слово встречается в тексте.
# Алгоритм:
# 1) Преобразует text к строке и разбивает её на слова методом split().
# 2) Проходит по словам в цикле.
# 3) Для каждого слова увеличивает счётчик в словаре частот.
# 4) Возвращает словарь частот.
def word_freq(text: str) -> dict:
    if text is None or not isinstance(text, str):
        return {}
    freq: dict[str, int] = {}
    words = str(text).split();
    for w in words:
        freq[w] = freq.get(w, 0) + 1
    return freq
````

````python
# Функционал:
# Группирует слова по их длине.
# Алгоритм:
# 1) Создаёт пустой словарь для групп.
# 2) Проходит по списку слов.
# 3) Для каждой длины создаёт список через setdefault и добавляет туда слово.
# 4) Возвращает словарь групп.
def group_words_by_len(words: list[str]) -> dict:
    out: dict[int, list[str]] = {}
    if words is None:
        return out
    for w in words:
        out.setdefault(len(w), []).append(w)
    return out
````

````python
# Функционал:
# Меняет местами ключи и значения словаря.
# Алгоритм:
# 1) Создаёт пустой словарь inv.
# 2) Проходит по всем парам key-value через items().
# 3) Записывает в новый словарь значение как ключ, а ключ как значение.
# 4) Возвращает инвертированный словарь.
def invert_dict(d: dict) -> dict:
    return {v: k for k, v in d.items()}
````

````python
# Функционал:
# Безопасно получает значение из вложенного словаря по пути из ключей.
# Алгоритм:
# 1) Начинает с корневого словаря d.
# 2) По очереди проходит по ключам из path.
# 3) Если текущий объект не dict или ключ отсутствует, возвращает default.
# 4) Иначе спускается на уровень глубже.
# 5) После прохода возвращает найденное значение.
from functools import reduce
def get_nested(d: dict, path: list[str], default=None):
    return reduce(
        lambda c, k:
            c.get(k, default) if isinstance(c, dict) else default, path, d)
````

````python
# Функционал:
# Строит индекс словарей по числовому id.
# Алгоритм:
# 1) Создаёт пустой словарь idx.
# 2) Проходит по списку элементов.
# 3) Берёт значение поля "id".
# 4) Если id является int и не является bool, сохраняет элемент в idx по этому id.
# 5) Возвращает индекс.
def build_index_by_id(items: list[dict]) -> dict:
    id_l = 'id'
    return {v: item for item in items 
        if isinstance(v := item.get(id_l), int) and not isinstance(v, bool)}
    
````

````python
# Функционал:
# Суммирует числовые значения по указанному ключу в списке словарей.
# Алгоритм:
# 1) Создаёт переменную total со значением 0.0.
# 2) Проходит по всем словарям в rows.
# 3) Получает значение по ключу key.
# 4) Если значение — int или float и не bool, добавляет его к total.
# 5) Возвращает сумму.
def sum_by_key(rows: list[dict], key: str) -> float:
    return sum(v for row in rows
        if isinstance((v := row.get(key)), (int, float)) and not isinstance(v, bool))
````

````python
# Функционал:
# Возвращает новую версию словаря item с обновлёнными полями из patch.
# Алгоритм:
# 1) Создаёт копию исходного словаря item.
# 2) Обновляет копию данными из patch через update.
# 3) Возвращает новый словарь, не изменяя исходный item.
def update_item_fields(item: dict, patch: dict) -> dict:
    if not item:
        return patch.copy()
    if not patch:
        return item.copy()    
    return item | patch
````

````python
# Демонстрация работы функций
# После заполнения TODO-частей запустите эту ячейку для быстрой самопроверки.

assert safe_get({"ab": 4}, "b") == None
assert safe_get({"ab": 4}, "a", "-1") == "-1"
assert safe_get({"ab": 4}, None) == None
assert safe_get(None, "a", "-1") == "-1"

assert merge_dicts({}, {}) == {}
assert merge_dicts({}, {"z": 99}) == {"z": 99}
assert merge_dicts({"z": 99}, {}) == {"z": 99}

a = {"id": 5}
result = merge_dicts(a, {"id": 6})
assert result is not a
assert merge_dicts({"p": 1, "q": 2}, {"q": 999, "r": 3}) == {"p": 1, "q": 999, "r": 3}

assert word_freq("Hello hello HELLO") == {"Hello": 1, "hello": 1, "HELLO": 1}
assert word_freq("") == {}
assert word_freq(None) == {}
assert word_freq(123) == {}
assert word_freq("python") == {"python": 1}
assert word_freq("a  b\nc\tc a") == {"a": 2, "b": 1, "c": 2}
assert word_freq("  one two two  ") == {"one": 1, "two": 2}
long_text = "foo bar " * 1000 + "foo"
result = word_freq(long_text)
assert result.get("foo") == 1001
assert result.get("bar") == 1000

assert word_freq("abc123 123 abc") == {"abc123": 1, "123": 1, "abc": 1}

assert group_words_by_len(["a", "bb", "c", "dd"]) == {1: ["a", "c"], 2: ["bb", "dd"]}
assert group_words_by_len([]) == {}
assert group_words_by_len(None) == {}
assert group_words_by_len(["hello"]) == {5: ["hello"]}
assert group_words_by_len(["a", "bc", "def"]) == {1: ["a"], 2: ["bc"], 3: ["def"]}
assert group_words_by_len(["cat", "dog", "bird"]) == {3: ["cat", "dog"], 4: ["bird"]}
assert group_words_by_len(["hi", "hi", "hello"]) == {2: ["hi", "hi"], 5: ["hello"]}

assert sum_by_key([{"a": False}, {"a": True}, {"a": 1}], "a") == 1.0

assert update_item_fields(None,{}) == {}
````

````python
# =========================
# LESSON 10 — TESTS (НЕ МЕНЯТЬ)
# =========================

def run_all_tests():
    it = make_item(1, "Hello", {"x": 1})
    assert it["id"] == 1 and it["title"] == "Hello" and it["payload"]["x"] == 1

    assert safe_get({"a": 1}, "a", 0) == 1
    assert safe_get({"a": 1}, "b", 0) == 0

    assert merge_dicts({"a": 1}, {"a": 2, "b": 3}) == {"a": 2, "b": 3}

    assert word_freq("a b a") == {"a": 2, "b": 1}
    assert word_freq("") == {}

    assert invert_dict({"a": 1, "b": 2}) == {1: "a", 2: "b"}

    data = {"a": {"b": {"c": 5}}}
    assert get_nested(data, ["a", "b", "c"]) == 5
    assert get_nested(data, ["a", "x"], default=0) == 0

    items = [{"id": 1, "title": "A"}, {"id": 2, "title": "B"}, {"id": "x", "title": "C"}]
    idx = build_index_by_id(items)
    assert idx[1]["title"] == "A" and idx[2]["title"] == "B"
    assert "x" not in idx

    rows = [{"a": 1}, {"a": 2.5}, {"b": 10}, {"a": "x"}]
    assert sum_by_key(rows, "a") == 3.5
    assert sum_by_key([], "a") == 0.0

    item2 = update_item_fields({"id": 1, "title": "A"}, {"title": "AA"})
    assert item2["title"] == "AA"
    assert {"id": 1, "title": "A"}["title"] == "A"  # исходный не меняем

    print("✅ LESSON 10: all tests passed")

run_all_tests()
````
