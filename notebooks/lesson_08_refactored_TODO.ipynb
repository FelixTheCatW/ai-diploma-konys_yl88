```python
# Функционал:
# Функция формирует список чисел от n до 0.
#
# Алгоритм:
# Запускаем цикл while, добавляем текущее число в список и уменьшаем его на 1, пока не дойдём до нуля.

def countdown(n: int) -> list[int]:
    out = []
    i = n
    while i >= 0:
        out.append(i)
        i -= 1
    return out
```

```python
# Функционал:
# Функция вычисляет сумму цифр неотрицательного целого числа.
#
# Алгоритм:
# В цикле while берём последнюю цифру через остаток от деления на 10, прибавляем её к сумме и отбрасываем цифру целочисленным делением.

def sum_digits(n: int) -> int:
    n = int(n)
    if n < 0:
        raise ValueError("n must be >= 0")    
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total
```

```python
# Функционал:
# Функция переворачивает цифры неотрицательного целого числа.
#
# Алгоритм:
# В цикле while по одной извлекаем последнюю цифру числа и собираем новое число справа налево.

def reverse_int(n: int) -> int:
    n = int(n)
    if n < 0:
        raise ValueError("n must be >= 0")
    rev = 0
    while n > 0:
        rev = rev * 10 + (n % 10)
        n //= 10
    return rev
```

```python
# Функционал:
# Функция считает количество делителей положительного числа.
#
# Алгоритм:
# Перебираем все числа от 1 до n в цикле while и увеличиваем счётчик, если число делится без остатка.

def count_divisors(n: int) -> int:
    n = int(n)
    if n <= 0:
        raise ValueError("n must be > 0")
    i = 1
    count = 0
    while i <= n:
        if  n % i == 0:
            count += 1
        i += 1
    return count
```

```python
# Функционал:
# Функция находит первую степень двойки, которая не меньше n.
#
# Алгоритм:
# Начинаем с 1 и в цикле while умножаем значение на 2, пока оно не станет больше или равно n.

def first_power_of_two_geq(n: int) -> int:
    n = int(n)
    if n <= 0:
        raise ValueError("n must be >= 0")
    i = 0
    while i < n:
        i = pow(2, i)
    return i
    
```

```python
# Функционал:
# Функция ищет первый элемент списка, который кратен числу k.
#
# Алгоритм:
# Проходим по списку по индексу в цикле while и сразу возвращаем первый подходящий элемент; если таких нет, возвращаем None.

def find_first_multiple(nums: list[int], k: int):
    if not nums:
        raise ValueError("nums cannot be empty")
    k = int(k)
    if k == 0:
        raise ValueError("k cannot be zero")    
    i = len(nums)
    n = 0
    while n < i:
        if nums[n] % k == 0:
            return nums[n]
        n += 1
    return None
```

```python
# Функционал:
# Функция удаляет отрицательные числа из списка и возвращает новый список.
#
# Алгоритм:
# В цикле while просматриваем все элементы и добавляем в результат только те, которые больше или равны нулю.

def remove_negatives(nums: list[int]) -> list[int]:
    if not nums:
        return []
    out = []
    i = 0
    while i < len(nums):
        if nums[i] >= 0:
            out.append(nums[i])
        i += 1
    return out 
```

```python
# Функционал:
# Функция повторяет строку, пока длина результата не достигнет target_len.
#
# Алгоритм:
# Создаём пустую строку и в цикле while дописываем исходную строку, пока длина результата меньше целевой.

def repeat_until_length(s: str, target_len: int) -> str:
    if not str:
        raise ValueError("input string cannot be empty")
    t = int(target_len)
    if t < 0:
        raise ValueError("target_len must be >= 0")
    ns = ""
    while len(ns) < t:
        ns += s        
    return ns
```

```python
# Функционал:
# Функция приближённо находит квадратный корень числа с шагом eps.
#
# Алгоритм:
# Начинаем с 0 и в цикле while увеличиваем значение на eps, пока следующий шаг ещё не превышает корень из n; затем округляем результат.
import math
def approx_sqrt(n: float, eps: float=0.001):
    if n < 0:
        raise ValueError("n must be >= 0")
    goal = 0
    while goal <= math.sqrt(n):
        goal += eps
    return round(goal, 1)    
```

```python
# Функционал:
# Функция извлекает целочисленный префикс из начала строки.
#
# Алгоритм:
# В цикле while двигаемся по строке, пока встречаем цифры; если цифр нет, возвращаем None, иначе преобразуем найденный префикс в int.

def parse_int_prefix(s: str):
    s = str(s)
    if not s:
        raise ValueError("s cannot be empty")
    i = 0    
    while i < len(s) and s[i].isnumeric():
        i += 1    

    return None if i == 0 else int(s[0:i])
```

```python
# =========================
# LESSON 08 — TESTS (НЕ МЕНЯТЬ)
# =========================

def run_all_tests():
    assert countdown(0) == [0]
    assert countdown(3) == [3, 2, 1, 0]

    assert sum_digits(0) == 0
    assert sum_digits(123) == 6

    assert reverse_int(0) == 0
    assert reverse_int(120) == 21
    assert reverse_int(123) == 321

    assert count_divisors(1) == 1
    assert count_divisors(6) == 4

    assert first_power_of_two_geq(1) == 1
    assert first_power_of_two_geq(9) == 16

    assert find_first_multiple([1, 3, 4, 8], 2) == 4
    assert find_first_multiple([1, 3, 5], 2) is None

    assert remove_negatives([1, -2, 0, 3]) == [1, 0, 3]
    assert remove_negatives([]) == []

    assert repeat_until_length("ab", 5) == "ababab"
    assert repeat_until_length("x", 1) == "x"
    assert repeat_until_length("x", 0) == ""

    assert approx_sqrt(4, 1e-3) == 2.0
    assert approx_sqrt(0, 1e-3) == 0.0

    assert parse_int_prefix("123abc") == 123
    assert parse_int_prefix("abc") is None
    assert parse_int_prefix("007x") == 7

    print("✅ LESSON 08: all tests passed")

run_all_tests()
```
